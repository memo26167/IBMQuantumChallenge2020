````python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
import numpy as np
from itertools import permutations
def week3_ans_func(problem_set):
    ##### build your quantum circuit here
    ##### In addition, please make it a function that can solve the problem even with different inputs (problem_set). We do validation with different inputs. 
    
    def boards_transform(problem_set):
        '''Transform the boards, from the coordinates to matrix full of ones'''
        boards = []
        for problem in problem_set:
            board = np.array([[0]*4]*4)
            for asteroid in problem:
                board[int(asteroid[0])][int(asteroid[1])] = 1
            boards.append(board.reshape(-1,))
        return boards
    
    def bits_permutations(n):
        '''Performs all possible permutations of n-bit string'''
        x = [] 
        for i in range(1<<n):
            s = bin(i)[2:]
            s = '0'*(n - len(s)) + s
            x.append(s)
        return x
    
    def qRAM(problem_set):
        '''Performs the QRAM operation,
        Using *boards_transform* and *bits_permutations* '''
        
        address = QuantumRegister(4, 'address')
        data = QuantumRegister(16, 'data')
        ancilla = QuantumRegister(3, 'ancilla')
        qc = QuantumCircuit(address, data, ancilla, name='QRAM')
        
        address_states = bits_permutations(4)
        boards = boards_transform(problem_set)
        for i, state in enumerate(address_states):
            board = boards[i]
            for j, bit in enumerate(state):
                if bit == '0':
                    qc.x(address[j])

            qc.rccx(address[0], address[1], ancilla[0])
            qc.rccx(address[2], address[3], ancilla[1])
            qc.rccx(ancilla[0], ancilla[1], ancilla[2])
            for j, asteroid in enumerate(board):
                if asteroid == 1:
                    qc.cx(ancilla[2], data[j])
            qc.rccx(ancilla[0], ancilla[1], ancilla[2])
            qc.rccx(address[2], address[3], ancilla[1])
            qc.rccx(address[0], address[1], ancilla[0])
            for j, bit in enumerate(state):
                if bit == '0':
                    qc.x(address[j])
            qc.barrier()
        return qc
    
    def oracle():
        data = QuantumRegister(16, 'data')
        flag = QuantumRegister(1, 'flag')
        solution_flag = QuantumRegister(1, 'solution_flag')
        columns_flag = QuantumRegister(1, 'columns_flag')
        twoconditions_flag = QuantumRegister(1, 'twoconditions_flag')
        ancilla = QuantumRegister(3, 'ancilla')
        qc = QuantumCircuit(data, flag,
                            solution_flag, columns_flag, twoconditions_flag,
                            ancilla, name='Oracle')
        def ccccx():
            control = QuantumRegister(4)
            target = QuantumRegister(1)
            ancilla = QuantumRegister(3)
            qc = QuantumCircuit(control, target, ancilla, name='cccx')
            qc.rccx(control[0], control[1], ancilla[0])
            qc.rccx(control[2], control[3], ancilla[1])
            qc.ccx(ancilla[0], ancilla[1],  target)
            qc.rccx(control[2], control[3], ancilla[1])
            qc.rccx(control[0], control[1], ancilla[0])
            return qc
            
        
        def check_twocolumnsequal(i, j):
            '''If columns i and j are equal, flip the columns_flag qubit'''
            qc.append(ccccx(), [data[0 + i], data[4  + i],  data[0 + j], data[4  + j]] 
                               + solution_flag[:] + ancilla[:])
            qc.append(ccccx(), [data[0 + i], data[8  + i],  data[0 + j], data[8  + j]]
                               + solution_flag[:] + ancilla[:])
            qc.append(ccccx(), [data[0 + i], data[12 + i],  data[0 + j], data[12 + j]]
                               + solution_flag[:] + ancilla[:])
            qc.append(ccccx(), [data[4 + i], data[8  + i],  data[4 + j], data[8  + j]]
                               + solution_flag[:] + ancilla[:])
            qc.append(ccccx(), [data[4 + i], data[12 + i],  data[4 + j], data[12 + j]]
                               + solution_flag[:] + ancilla[:])
            qc.append(ccccx(), [data[8 + i], data[12 + i],  data[8 + j], data[12 + j]]
                               + solution_flag[:] + ancilla[:])
        
        # check if the column 0 and column 1 are equal
        check_twocolumnsequal(0, 1)
        check_twocolumnsequal(0, 2)
        check_twocolumnsequal(0, 3)
        check_twocolumnsequal(1, 2)
        check_twocolumnsequal(1, 3)
        check_twocolumnsequal(2, 3)
        
        def check_unsolvability():
            '''If data qubits 0, 5, 10 and 15 are 1s, we flip the solution_flag qubit'''
            qc.append(ccccx(), [data[0], data[5], data[10], data[15]] + solution_flag[:] + ancilla[:])
            '''If both, solution_flag qubit and columns_flag are 1s, then flip twoconditions_flag'''
            qc.ccx(solution_flag, columns_flag, twoconditions_flag)
            '''If twoconditions_flag is 1, flip solution_flag, 
               this way, if two columns are equal and is a solution, 
               we avoid the problem of the standard flag being flipped two times'''
            qc.cx(twoconditions_flag, solution_flag)
        
        # check if the puzzle is unsolvable
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])

        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        #####

        check_unsolvability()
        # swap column 0 with 1
        for i in range(4):
            qc.swap(data[0 + 4*i], data[1 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        #####
        
        check_unsolvability()
        # swap column 0 with 2
        for i in range(4):
            qc.swap(data[0 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 2
        for i in range(4):
            qc.swap(data[1 + 4*i], data[2 + 4*i])
                
        #####
        
        check_unsolvability()
        # swap column 0 with 2
        for i in range(4):
            qc.swap(data[0 + 4*i], data[2 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 3
        for i in range(4):
            qc.swap(data[1 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 1 with 3
        for i in range(4):
            qc.swap(data[1 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 2 with 3
        for i in range(4):
            qc.swap(data[2 + 4*i], data[3 + 4*i])
        
        check_unsolvability()
        # swap column 0 with 3
        for i in range(4):
            qc.swap(data[0 + 4*i], data[3 + 4*i])
        
        '''If solution_flag is 1, flip flag to mark it'''
        qc.cx(solution_flag, flag)
        return qc
    
    def diffuser():
        diffuss = QuantumRegister(4, 'diffuss')
        ancilla = QuantumRegister(3, 'mct_ancilla')
        qc = QuantumCircuit(diffuss, ancilla, name='Diffusser')
        # Applying hadamard to all Qubits to be diffused
        qc.h(diffuss[:])
        # Applying the Uf0 operator
        qc.x(diffuss[:])
        qc.h(diffuss[-1])
        # Controlled, flip last qubit of diffuss
        qc.rccx(diffuss[0], diffuss[1], ancilla[0])
        qc.ccx(ancilla[0], diffuss[2], diffuss[3])
        qc.rccx(diffuss[0], diffuss[1], ancilla[0])
        qc.h(diffuss[-1])
        qc.x(diffuss[:])
        #applying 2nd hadamard to all qubits
        qc.h(diffuss[:])
        return qc
        
    # Initialize Quantum Circuit
    address = QuantumRegister(4, 'address')
    data = QuantumRegister(16, 'data')
    
    flag = QuantumRegister(1, 'flag')
    solution_flag = QuantumRegister(1, 'solution_flag')
    columns_flag = QuantumRegister(1, 'columns_flag')
    twoconditions_flag = QuantumRegister(1, 'twoconditions_flag')
    
    ancilla = QuantumRegister(3, 'ancilla')
    cr = ClassicalRegister(4)
    qc = QuantumCircuit(address, data, flag, 
                        solution_flag, columns_flag, twoconditions_flag, 
                        ancilla, cr)
    
    # Initialize address
    qc.h(address[:])
    qc.barrier()
    
    # Initialize flag
    qc.x(flag[:])
    qc.h(flag[:])
    qc.barrier()
    
    #### Code for Grover's algorithm with iterations = 1
    for i in range(1):
        # Initialize QRAM
        qc.append(qRAM(problem_set), address[:] + data[:] + ancilla[:])
        # Perform Oracle 
        qc.append(oracle(), data[:] + flag[:] + 
                  solution_flag[:] + columns_flag[:] + twoconditions_flag[:]
                  + ancilla[:])
        # Undo QRAM
        qc.append(qRAM(problem_set).inverse(), address[:] + data[:] + ancilla[:])
        # Apply Diffuser
        qc.append(diffuser(), address[:] + ancilla[:])

    qc.measure(address[:], cr[:])
    return qc
````
